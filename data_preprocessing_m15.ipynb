```python
import pandas as pd
import sqlite3

# 데이터베이스 파일 경로
db_file = 'btc_data_m15_20170901_20241205.db'

# 데이터베이스 연결
con = sqlite3.connect(db_file)

# SQL 쿼리 실행 및 데이터프레임으로 불러오기
query = "SELECT * FROM ticker_data"
df = pd.read_sql_query(query, con)

# 데이터베이스 연결 종료
con.close()

# 데이터프레임 출력
df.head()  # 데이터프레임의 첫 5행 출력
```

```python
df
```

```python
df = df.drop_duplicates(keep='first')
df
```

```python
df.info()
```

```python
# trade_date와 trade_time을 결합하여 새로운 datetime 컬럼 생성
df['datetime'] = pd.to_datetime(df['trade_date'] + ' ' + df['trade_time'])

# 결과 출력
df.head()
```

```python
df.info()
```

```python
df_sorted = df.sort_values(by='datetime')
df_sorted
```

```python
import sqlite3

# SQLite 데이터베이스 파일 경로
db_file = "btc_data_m15_sorted.db"

# 데이터베이스 연결
con = sqlite3.connect(db_file)

# 데이터프레임을 SQLite 데이터베이스에 저장
df_sorted.to_sql(name='ticker_data', con=con, if_exists='replace', index=False)

# 연결 종료
con.close()

print("데이터프레임이 SQLite 데이터베이스에 저장되었습니다.")
```

```python
df = df_sorted
df
```

```python
df['volatility'] = df['high_price'] - df['low_price']
```

```python
df
```

```python
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df
```

```python
# 다음 봉의 상승률 계산
df['next_return'] = df['trade_price'].shift(-1) / df['trade_price'] - 1

# 변동성 돌파 성공률 계산
df['target'] = df['opening_price'] + (df['high_price'].shift(1) - df['low_price'].shift(1)) * 0.5
df['success'] = (df['high_price'] > df['target']).astype(int)
df
```

```python
df.rename(columns={'trade_price':'closing_price'}, inplace=True)
df
```

```python
new_column_order = [
    'datetime', 'type', 'code', 'opening_price', 'closing_price', 'high_price', 'low_price', 'trade_volume', 'volatility', 'next_return', 'target', 'success', 'trade_date', 'trade_time', 'hour', 'minute', 'trade_timestamp', 
]

df = df[new_column_order] 
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
import sqlite3

# SQLite 데이터베이스 파일 경로
db_file = "btc_data_m15_preprocessing_ex.db"

# 데이터베이스 연결
con = sqlite3.connect(db_file)

# 데이터프레임을 SQLite 데이터베이스에 저장
df.to_sql(name='ticker_data', con=con, if_exists='replace', index=False)

# 연결 종료
con.close()

print("데이터프레임이 SQLite 데이터베이스에 저장되었습니다.")
```
