7. prev_closing_price: 
    - 직전 거래일의 종가.
8. acc_trade_price: 
    - 누적 거래 금액.
9. change: 
    - 가격 변화 상태 (예: 상승, 하락, 보합).
10. RISE:

11. change_price: 
    - 가격 변화량.
12. signed_change_price:

13. change_rate: 
    - 가격 변화율.
14. signed_change_rate: 
    - 변화율의 부호가 포함된 가격 변화율.
15. ask_bid: 
    - 매도/매수 정보.
16. trade_volume: 
    - 해당 기간의 거래량.
17. acc_trade_volume: 
    - 누적 거래량.

    - 거래 날짜.
19. trade_time: 
    - 거래 시간.
20. trade_timestamp:

21. acc_ask_volume: 
    - 누적 매도량.
22. acc_bid_volume: 
    - 누적 매수량.
23. highest_52_week_price: 
    - 최근 52주 동안의 최고 가격.
24. highest_52_week_date: 
    - 최고 가격이 발생한 날짜.
25. lowest_52_week_price:

29. delisting_date:

30. market_warning:

31. timestamp: 
    - 데이터 수집 시점의 타임스탬프.
32. acc_trade_price_24h: 
    - 24시간 누적 거래 금액.
33. acc_trade_volume_24h: 
    - 24시간 누적 거래량.
34. stream_type: 
    - 데이터 스트림 유형 (예: 실시간 데이터).

1. type:
    - orderbook: 매수/매도 호가 정보를 포함한 데이터
2. code:
    - 거래되는 암호화폐의 코드로, 여기서는 원화(KRW)로 거래되는 비트코인(BTC)을 나타냅니다.
3. timestamp:
    - 데이터 수신 시점의 타임스탬프입니다. 일반적으로 UTC 기준의 밀리초 단위로 표현됩니다.
4. total_ask_size:
    - 현재 전체 매도(ask) 주문의 총량을 나타냅니다. 이는 현재 시장에서 매도할 수 있는 총 비트코인 수량입니다.
5. total_bid_size:
    - 현재 전체 매수(bid) 주문의 총량을 나타냅니다. 이는 현재 시장에서 매수할 수 있는 총 비트코인 수량입니다.
6. orderbook_units:
    - 주문서의 각 단위를 포함하는 리스트입니다. 각 단위는 다음과 같은 키를 포함합니다:
    - ask_price: 매도 가격. 이 가격에 매도 주문이 대기하고 있습니다.
    - bid_price: 매수 가격. 이 가격에 매수 주문이 대기하고 있습니다.
    - ask_size: 해당 매도 가격에서의 매도 수량.
    - bid_size: 해당 매수 가격에서의 매수 수량.
    - 이 리스트는 여러 개의 주문서 단위를 포함하며, 각 단위는 매도 및 매수의 가격과 수량 정보를 제공합니다.
7. stream_type:
    - 'REALTIME': 데이터 스트림의 유형을 나타내며, 실시간으로 수신되는 데이터임을 의미합니다.
8. level:
    -  주문서의 레벨을 나타냅니다. 레벨 0은 일반적으로 전체 주문서 정보를 나타내며, 더 높은 레벨은 특정 깊이의 정보(예: 상위 매도/매수 주문만)를 표시할 수 있습니다.

1. type:
    - trade: 개별 거래 정보
2. code:
    - 거래되는 암호화폐의 코드로, 여기서는 원화(KRW)로 거래되는 비트코인(BTC)을 나타냅니다.
3. timestamp:
    - 데이터 수신 시점의 타임스탬프입니다. 일반적으로 UTC 기준의 밀리초 단위로 표현됩니다.
4. trade_date:
    - 거래가 발생한 날짜를 나타냅니다.
5. trade_time:
    - 거래가 발생한 시간을 나타냅니다.
6. trade_timestamp:
    - 거래가 발생한 시점의 타임스탬프입니다. 일반적으로 UTC 기준의 밀리초 단위로 표현됩니다.
7. trade_price:
    - 해당 거래의 가격, 즉 비트코인이 거래된 가격입니다.
8. trade_volume:
    - 해당 거래에서 거래된 비트코인의 수량입니다. 이 경우 약 0.00006652 비트코인이 거래되었습니다.
9. ask_bid:
    - 거래의 유형을 나타내며, 여기서는 매수(bid) 주문에 의해 체결된 거래임을 의미합니다. 반대로 매도 주문에 의해 체결된 경우는 'ASK'로 표시됩니다.
10. prev_closing_price:
    - 직전 거래일의 종가입니다. 이는 이전 거래 종료 시점의 비트코인 가격입니다.
11. change:
    - 가격 변화 상태를 나타내며, 여기서는 가격이 상승했음을 의미합니다.
12. change_price:
    - 가격 변화량을 나타냅니다. 즉, 현재 가격이 이전 가격보다 1,578,000원 상승했음을 의미합니다.
13. sequential_id:
    - 거래의 고유 식별자입니다. 이 값은 각 거래에 대해 고유하게 부여되며, 거래의 순서를 추적하는 데 사용됩니다.
14. best_ask_price:
    - 현재 가장 낮은 매도 가격, 즉 시장에서 가장 먼저 체결될 수 있는 매도 가격입니다.
15. best_ask_size:
    - 현재 가장 낮은 매도 가격에서 매도 가능한 비트코인의 수량입니다.
16. best_bid_price:
    - 현재 가장 높은 매수 가격, 즉 시장에서 가장 먼저 체결될 수 있는 매수 가격입니다.
17. best_bid_size:
    - 현재 가장 높은 매수 가격에서 매수 가능한 비트코인의 수량입니다.
18. stream_type:
    - 데이터 스트림의 유형을 나타내며, 실시간으로 수신되는 데이터임을 의미합니다.

```python
import multiprocessing as mp
import pyupbit
import datetime
import pandas as pd
import sqlite3


def create_table(con):
    cursor = con.cursor()
    # 테이블 생성 쿼리
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS ticker_data (
        type TEXT,
        code TEXT,
        opening_price REAL,
        high_price REAL,
        low_price REAL,
        trade_price REAL,
        prev_closing_price REAL,
        acc_trade_price REAL,
        change TEXT,
        change_price REAL,
        signed_change_price REAL,
        change_rate REAL,
        signed_change REAL,
        acc_bid_volume REAL,
        highest_52_week_price REAL,
        highest_52_week_date TEXT,
        lowest_52_week_price REAL,
        lowest_52_week_date TEXT,
        market_state TEXT,
        is_trading_suspended BOOLEAN,
        delisting_date TEXT,
        market_warning TEXT,
        timestamp INTEGER,
        acc_trade_price_24h REAL,
        acc_trade_volume_24h REAL,
        stream_type TEXT_rate REAL,
        ask_bid TEXT,
        trade_volume REAL,
        acc_trade_volume REAL,
        trade_date TEXT,
        trade_time TEXT,
        trade_timestamp INTEGER,
        acc_ask_volume
    )
    ''')
    con.commit()


def save_data(con, ticker_data):
    columns = [
        'type', 'code', 'opening_price', 'high_price', 'low_price',
        'trade_price', 'prev_closing_price', 'acc_trade_price', 
        'change', 'change_price', 'signed_change_price', 
        'change_rate', 'signed_change_rate', 'ask_bid', 
        'trade_volume', 'acc_trade_volume', 'trade_date', 
        'trade_time', 'trade_timestamp', 'acc_ask_volume', 
        'acc_bid_volume', 'highest_52_week_price', 
        'highest_52_week_date', 'lowest_52_week_price', 
        'lowest_52_week_date', 'market_state', 
        'is_trading_suspended', 'delisting_date', 
        'market_warning', 'timestamp', 
        'acc_trade_price_24h', 'acc_trade_volume_24h', 'stream_type'
    ]

    for k, v in ticker_data.items():
        if len(v) != 0:
            df = pd.DataFrame(data=v, columns=columns)
            try:
                df.to_sql(name='ticker_data', con=con, if_exists='append', index=False, chunksize=1000)
                print(f"Data for {k} saved successfully.")
            except Exception as e:
                print(f"Error saving data for {k}: {e}")


if __name__ == "__main__":
    krw_tickers = pyupbit.get_tickers(fiat="KRW")
    queue = mp.Queue()
    proc = mp.Process(
        target=pyupbit.WebSocketClient,
        args=('ticker', krw_tickers, queue),
        daemon=True
    )
    proc.start()

    ticker_data = {k: [] for k in krw_tickers}
    data_count = 0
    data_count_max = 10000
    con = sqlite3.connect("./2022-06-29-coin.db")

    # 테이블 생성
    create_table(con)

    while True:
        data = queue.get()
        data_count += 1

        # 모든 컬럼을 변수로 선언
        type_ = data['type']
        code = data['code']
        opening_price = data['opening_price']
        high_price = data['high_price']
        low_price = data['low_price']
        trade_price = data['trade_price']
        prev_closing_price = data['prev_closing_price']
        acc_trade_price = data['acc_trade_price']
        change = data['change']
        change_price = data['change_price']
        signed_change_price = data['signed_change_price']
        change_rate = data['change_rate']
        signed_change_rate = data['signed_change_rate']
        ask_bid = data['ask_bid']
        trade_volume = data['trade_volume']
        acc_trade_volume = data['acc_trade_volume']
        trade_date = data['trade_date']
        trade_time = data['trade_time']
        trade_timestamp = data['trade_timestamp']
        acc_ask_volume = data['acc_ask_volume']
        acc_bid_volume = data['acc_bid_volume']
        highest_52_week_price = data['highest_52_week_price']
        highest_52_week_date = data['highest_52_week_date']
        lowest_52_week_price = data['lowest_52_week_price']
        lowest_52_week_date = data['lowest_52_week_date']
        market_state = data['market_state']
        is_trading_suspended = data['is_trading_suspended']
        delisting_date = data['delisting_date']
        market_warning = data['market_warning']
        timestamp = data['timestamp']
        acc_trade_price_24h = data['acc_trade_price_24h']
        acc_trade_volume_24h = data['acc_trade_volume_24h']
        stream_type = data['stream_type']

        # 모든 컬럼을 포함한 row 생성
        row = (
            type_, code, opening_price, high_price, low_price,
            trade_price, prev_closing_price, acc_trade_price, 
            change, change_price, signed_change_price, 
            change_rate, signed_change_rate, ask_bid, 
            trade_volume, acc_trade_volume, trade_date, 
            trade_time, trade_timestamp, acc_ask_volume, 
            acc_bid_volume, highest_52_week_price, 
            highest_52_week_date, lowest_52_week_price, 
            lowest_52_week_date, market_state, 
            is_trading_suspended, delisting_date, 
            market_warning, timestamp, 
            acc_trade_price_24h, acc_trade_volume_24h, stream_type
        )

        # ticker_data에 추가
        ticker_data[code].append(row)

        print(data_count)
        if data_count == data_count_max:
            print("*" * 80)
            save_data(con, ticker_data)
            ticker_data = {k: [] for k in krw_tickers}
            data_count = 0
```

```python
import os
import pyupbit
import datetime
import pandas as pd
import sqlite3

def create_table(con):
    cursor = con.cursor()
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS ticker_data (
        type TEXT,
        code TEXT,
        opening_price REAL,
        high_price REAL,
        low_price REAL,
        trade_price REAL,
        trade_volume REAL,
        trade_date TEXT,
        trade_time TEXT,
        trade_timestamp INTEGER
    )
    ''')
    con.commit()

def save_data(con, df):
    try:
        df.to_sql(name='ticker_data', con=con, if_exists='append', index=False)
        print("Data saved successfully.")
    except Exception as e:
        print(f"Error saving data: {e}")

def fetch_data(ticker, start, end):
    df = pyupbit.get_ohlcv(ticker, interval='minute1', to=end.strftime('%Y-%m-%d %H:%M:%S'), count=2000)
    df = df[(df.index >= start) & (df.index < end)]
    return df

if __name__ == "__main__":
    ticker = "KRW-BTC"  # 원하는 암호화폐 티커

    
    # 2024년 12월 3일의 시작과 종료 시간 설정
    start_date = datetime.datetime(2024, 12, 4, 0, 0)
    end_date = datetime.datetime(2024, 12, 5, 0, 0)

    db_file = "btc_data_m_2024_12_04.db"
    
    # 기존 데이터베이스 파일 삭제 (새로 생성하기 위해)
    if os.path.exists(db_file):
        os.remove(db_file)

    con = sqlite3.connect(db_file)
    create_table(con)

    # 데이터 fetch
    df = fetch_data(ticker, start_date, end_date)

    if df is not None and not df.empty:
        df['type'] = 'ticker'  # 타입 추가
        df['code'] = ticker  # 코드 추가
        df['trade_timestamp'] = df.index.astype('int64') // 10**9  # 타임스탬프 추가 (초 단위)
        
        # 날짜와 시간 컬럼 추가
        df['trade_date'] = df.index.date.astype(str)  # YYYY-MM-DD 형식
        df['trade_time'] = df.index.strftime('%H:%M')  # HH:MM 형식

        # 데이터베이스에 저장
        save_data(con, df[['type', 'code', 'open', 'high', 'low', 'close', 'volume', 'trade_date', 'trade_time', 'trade_timestamp']].rename(columns={
            'open': 'opening_price',
            'high': 'high_price',
            'low': 'low_price',
            'close': 'trade_price',
            'volume': 'trade_volume'
        }))
    else:
        print("No data found for the specified date.")

    con.close()
```

```python
import os
import pyupbit
import datetime
import pandas as pd
import sqlite3
import time

def create_table(con):
    cursor = con.cursor()
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS ticker_data (
        type TEXT,
        code TEXT,
        opening_price REAL,
        high_price REAL,
        low_price REAL,
        trade_price REAL,
        trade_volume REAL,
        trade_date TEXT,
        trade_time TEXT,
        trade_timestamp INTEGER
    )
    ''')
    con.commit()

def save_data(con, df):
    try:
        df.to_sql(name='ticker_data', con=con, if_exists='append', index=False)
        print("Data saved successfully.")
    except Exception as e:
        print(f"Error saving data: {e}")

def fetch_data(ticker, start, end):
    all_data = []
    current = start
    while current < end:
        next_current = min(current + datetime.timedelta(minutes=200), end)
        to_time = next_current - datetime.timedelta(hours=9)  # UTC로 변환
        df = pyupbit.get_ohlcv(ticker, interval='minute1', to=to_time.strftime('%Y-%m-%d %H:%M:%S'), count=200)
        if df is not None and not df.empty:
            all_data.append(df)
        current = next_current
        time.sleep(0.1)  # API 호출 제한 방지
    
    if all_data:
        combined_df = pd.concat(all_data)
        return combined_df[(combined_df.index >= start) & (combined_df.index < end)]
    return pd.DataFrame()

if __name__ == "__main__":
    ticker = "KRW-BTC"  # 원하는 암호화폐 티커
    start_date = datetime.datetime(2024, 11, 25, 0, 0)
    end_date = datetime.datetime(2024, 12, 2, 0, 0)  # 12월 1일까지 포함하기 위해

    db_file = "btc_data_m_20241125_20241201.db"

    # 기존 데이터베이스 파일 삭제 (새로 생성하기 위해)
    if os.path.exists(db_file):
        os.remove(db_file)

    con = sqlite3.connect(db_file)
    create_table(con)

    current_date = start_date
    while current_date < end_date:
        next_date = min(current_date + datetime.timedelta(days=1), end_date)

        # 데이터 fetch
        df = fetch_data(ticker, current_date, next_date)
        if df is not None and not df.empty:
            df['type'] = 'ticker'  # 타입 추가
            df['code'] = ticker  # 코드 추가
            df['trade_timestamp'] = df.index.astype('int64') // 10**9  # 타임스탬프 추가 (초 단위)
            # 날짜와 시간 컬럼 추가
            df['trade_date'] = df.index.date.astype(str)  # YYYY-MM-DD 형식
            df['trade_time'] = df.index.strftime('%H:%M')  # HH:MM 형식

            # 데이터베이스에 저장
            save_data(con, df[['type', 'code', 'open', 'high', 'low', 'close', 'volume', 'trade_date', 'trade_time', 'trade_timestamp']].rename(columns={
                'open': 'opening_price',
                'high': 'high_price',
                'low': 'low_price',
                'close': 'trade_price',
                'volume': 'trade_volume'
            }))
        else:
            print(f"No data found for {current_date.date()}")

        print(f"Completed processing for {current_date.date()}")

        current_date = next_date
        time.sleep(1)  # API 요청 제한을 피하기 위한 대기 시간

    con.close()
    print("Data collection completed for all dates.")
```
