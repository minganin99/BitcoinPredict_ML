```python
def get_dynamic_multiplier(hour):
    # 시간대별 최적 multiplier 설정
    if hour == 9:  # 오전 9시
        return 0.3
    elif hour in [21, 22, 23, 0]:  # 저녁 9시-자정
        return 0.5
    elif hour in [2, 3, 4]:  # 새벽 2-4시
        return 0.7
    elif hour in [10, 11]:  # 오전 10-11시
        return 0.4
    elif hour in [13, 14, 15, 16]:  # 오후 1-4시
        return 0.7
    elif hour in [17, 18, 19]:  # 오후 5-7시
        return 0.6
    else:  # 나머지 시간대
        return 0.5
```

```python
df['volatility'] = df['high_price'] - df['low_price']
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute

# 다음 봉의 상승률 계산
df['next_return'] = df['trade_price'].shift(-1) / df['trade_price'] - 1

# 시간대별 동적 multiplier 적용
df['multiplier'] = df['hour'].map(get_dynamic_multiplier)

# 변동성 돌파 성공률 계산 (동적 multiplier 적용)
df['target'] = df['opening_price'] + (df['high_price'].shift(1) - df['low_price'].shift(1)) * df['multiplier']
df['success'] = (df['high_price'] > df['target']).astype(int)

```

```python
df.rename(columns={'trade_price':'closing_price'}, inplace=True)
new_column_order = [
    'datetime', 'type', 'code', 'opening_price', 'closing_price', 'high_price', 'low_price', 'trade_volume', 'volatility', 'next_return', 'target', 'success', 'multiplier', 'trade_date', 'trade_time', 'hour', 'minute', 'trade_timestamp', 
]
df = df[new_column_order] 
df.reset_index(drop=True, inplace=True)
```

```python
import sqlite3

# SQLite 데이터베이스 파일 경로
db_file = "btc_data_m1_preprocessing_ex_add.db"

# 데이터베이스 연결
con = sqlite3.connect(db_file)

# 데이터프레임을 SQLite 데이터베이스에 저장
df.to_sql(name='ticker_data', con=con, if_exists='replace', index=False)

# 연결 종료
con.close()

print("데이터프레임이 SQLite 데이터베이스에 저장되었습니다.")
```
